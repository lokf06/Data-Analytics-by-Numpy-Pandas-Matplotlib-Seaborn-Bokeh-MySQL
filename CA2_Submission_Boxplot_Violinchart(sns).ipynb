```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv("ResaleflatpricesbasedonregistrationdatefromJan2017onwards.csv")
df.head()
```

```python
### Text-based Analysis using Pandas ###

# Display shape of the dataset
print("<< This is the shape of the dataset >>")
print(df.shape)
print()

# Display index range
print("<< This is the index of the dataset >>")
print(df.index)
print()

# Display all columns in the dataset
print("<< These are the columns in the dataset >>")
print(df.columns)
print()

# Display summary of the dataset
print("<< A summary of this dataset is shown below >>")
print(df.info())
print()

# Count total number of non-NA (non-null) values in the dataset
print("<< The total number of non-NA values in this dataset >>")
print(df.count())
print()

# Display statistical summary of the quantitative variables columns
print("<< A descriptive statistical summary of the quantitative data in this dataset >>")
df[['resale_price', 'floor_area_sqm']].describe()
```

```python
# Drop rows where 'resale_price' has 0 value
df = df[df['resale_price']!= 0]
```

```python
df.info()
```

```python
df.head()
```

```python
### Plotting Boxplot and Median Table using Seaborn ###
# Filter the DataFrame for the desired flat types and towns
flat_types = ["2 ROOM", "3 ROOM", "4 ROOM", "5 ROOM"]
towns = ["JURONG EAST", "ANG MO KIO", "TAMPINES"]
filtered_df = df[df['flat_type'].isin(flat_types) & df['town'].isin(towns)]

# Create the boxplot
plt.figure(figsize=(16, 12))
palette = sns.color_palette("Set1", n_colors=len(towns))
ax = sns.boxplot(data=filtered_df, x='flat_type', y='resale_price', hue='town', palette=palette)
plt.title("Resale Prices for Different Flat Types and Towns")
plt.xlabel("Flat Type")
plt.ylabel("Resale Price")

# Calculate median values
medians = filtered_df.groupby(['flat_type', 'town'])['resale_price'].median().unstack()

# Prepare data for the table
table_data = medians.reset_index()
table_data.columns = ['Flat Type'] + towns
table_data = table_data.set_index('Flat Type')

# Add a table to the plot
table = plt.table(cellText=table_data.values,
                  rowLabels=table_data.index,
                  colLabels=table_data.columns,
                  cellLoc='center',
                  loc='bottom',
                  bbox=[0.1, -0.5, 0.8, 0.3])  

# Add title for the table
plt.text(0.5, -0.55, 'Median Resale Prices by Flat Type and Town', ha='center', va='center', fontsize=12, weight='bold', 
         transform=plt.gca().transAxes)

# Adjust the layout to make space for the table
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.35)

# Create a legend with town names
town_legend_patches = [Patch(facecolor=palette[i], edgecolor="black", label=town) for i, town in enumerate(towns)]
plt.legend(handles=town_legend_patches, title="Town", loc="upper right")

plt.show()
```

```python
### Plotting Violin Chart using seaborn ###

# Filter the DataFrame for the desired flat types and towns
flat_types = ["2 ROOM", "3 ROOM", "4 ROOM", "5 ROOM"]
towns = ["JURONG EAST", "ANG MO KIO", "TAMPINES"]
filtered_df = df[df['flat_type'].isin(flat_types) & df['town'].isin(towns)]

# Create the violin plot
plt.figure(figsize=(16, 10))
palette = sns.color_palette("Set1", n_colors=len(towns))
ax = sns.violinplot(data=filtered_df, x='flat_type', y='resale_price', hue='town', palette=palette)
plt.title("Resale Prices for Different Flat Types and Towns")
plt.xlabel("Flat Type")
plt.ylabel("Resale Price")

# Create legend with town names
town_legend_patches = [Patch(facecolor=palette[i], edgecolor="black", label=town) for i, town in enumerate(towns)]
plt.legend(handles=town_legend_patches, title="Town", loc="upper right")

plt.show()
```
